```python
obs = sess.graph.get_tensor_by_name("default_policy/observation:0")
nodes = [n.name for n in tf.get_default_graph().as_graph_def().node]

def compute_tensor(node):
    try:
        return sess.run(sess.graph.get_tensor_by_name(node+":0"), {obs:[[0.22134309, 0.00360639, 0.06868469]]})
    except:
        return None

results = {}
for node in nodes:
    results[node] = compute_tensor(node)

print(results)
```

```python
import numpy as np

for k, v in results.items():
    try:
        if (v is not None) and (isinstance(v, np.ndarray) or isinstance(v, list)) and v.size == 1 and np.isscalar(v[0]):
            if np.isclose(v[0], 0.0012322, atol=1e-7, rtol=1e-3):
                print(k)
                print("===========")
    except:
        print("========")
        print(k)
        print(v)
```
